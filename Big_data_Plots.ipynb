# Plotting



```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
execution_times = [
    [95, 91, 86],  # Dataset 1 GB
    [917, 1162, 1129],    # Dataset 12 GB
    [1877, 2340, 2439],  # Dataset 24 GB
    [4922, 4788, 4836],     # Dataset 48 GB
]

# Dataset labels
dataset_labels = ['1 GB', '12 GB', '24 GB', '48 GB']

# Number of datasets and executions
num_datasets = len(execution_times)
num_executions = len(execution_times[0])
```

```python
# Plotting
fig, ax = plt.subplots()

# Creating x positions for each dataset
x = np.arange(num_datasets)

# Plotting the execution times for each execution as individual bars
width = 0.3  # Width of the bars

for i in range(num_executions):
    execution_time = [execution_times[j][i] for j in range(num_datasets)]
    bars = ax.bar(x + i * width, execution_time, width, label=f'Execution {i+1}')

    # Annotate each bar with its value
    for bar, value in zip(bars, execution_time):
        ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.1,
                f'{value}', ha='center', va='bottom')

# Adding labels and title
ax.set_xlabel('Dataset Size')
ax.set_ylabel('Execution Time (s)')
ax.set_title('Execution Times for Different Datasets')
ax.set_xticks(x + width * num_executions / 2)
ax.set_xticklabels(dataset_labels)
ax.legend()

# Displaying the plot
plt.show()
```

```python
# Calculate the average execution time for each dataset
averages = [np.mean(times) for times in execution_times]

# Print the averages
for label, avg in zip(dataset_labels, averages):
    print(f'Average execution time for {label}: {avg:.2f} seconds')
```
